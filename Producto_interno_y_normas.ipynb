**Podemos expresar la norma L2 al cuadrado como el producto interno entre nuestra matriz y su transpuesta. Esto tiene una serie de ventajas computacionales**

**Veamos que podemos expresar al producto interno de dos vectores como la norma (Es decir la norma L2) de cada vector multiplicado por el coseno del angulo que están formando**

producto interno en normas.PNG

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v1 = np.array([0,0,0,3]) #Vector que incluya el origen directamente
v2 = np.array([0,0,3,3])
```

```python
plt.xlim(-2,6)
plt.ylim(-2,6)

plt.quiver([v1[0], v2[0]],
           [v1[1], v2[1]],
           [v1[2], v2[2]],
           [v1[3], v2[3]],
           angles = "xy",
           scale_units = "xy",
           scale = 1,
           color = sns.color_palette()
           )

plt.show()
```

Nuestros dos vectores forman un cuadrado. un cuadrado tiene todos sus lados de 90 grados. Por lo que, al dividir mi cuadrado en dos triangulos mediante el vector v2, el angulo formado entre ambos vectores será de 45 grados. 

angulo_45.PNG

Probemos la igualdad entonces manifestada en el primer apartado:

```python
v1 = np.array([0,3]) # Reescribo mis vectores sin poner el origen 0,0
v2 = np.array([3,3])

print(v1.T.dot(v2))
```

```python
norma_v1 = np.linalg.norm(v1)
norma_v2 = np.linalg.norm(v2)

print(norma_v1 * norma_v2 * np.cos(np.deg2rad(45))) # la función coseno trabaja con radianes. Y
# el angulo que acabo de calcular está en grados. Entonces vamos a usar la función
# np.deg2rad(cant de grados)
```
