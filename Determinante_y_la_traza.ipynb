El determinante de una matriz nos habla sobre la transformacion que ejerce una matriz sobre el espacio que está transformando. Por ej si estamos en r2 y el determinante es negativo nos está danto el espejo de nuestro espacio.

```python
import matplotlib.pyplot as plt
```

```python
def graficarVectores(vecs,cols,alpha=1):
  plt.figure()
  plt.axvline(x=0,color="grey",zorder=0)
  plt.axhline(y=0,color="grey",zorder=0)

  for i in range(len(vecs)):
    x = np.concatenate([[0,0], vecs[i]])
    plt.quiver([x[0]],
               [x[1]],
               [x[2]],
               [x[3]],
               angles= "xy", scale_units="xy",scale=1,
                color=cols[i],
               alpha = alpha)
```

```python
v1 = np.array([0,1])
v2 = np.array([1,0])

graficarVectores([v1,v2],["blue","red"])

plt.xlim(-0.25,2)
plt.ylim(-0.25,2)

plt.show()
```

```python
A = np.array([[2,0],[0,2]])
print(A)
```

Esta matriz solo amplifica las coordenadas. Es decir solo duplicaría la distancia de mis dos vectores anteriores. veamoslo

```python
v1_transformado = A.dot(v1)
v2_transformado = A.dot(v2)

```

```python
graficarVectores([v1_transformado,v2_transformado],["orange","green"])

plt.xlim(-0.25,3)
plt.ylim(-0.25,3)

plt.show()
```

Cada vector se quedo donde estaba pero se alargo. El area de nuestro vectores cuando median 1 era 1 mientras que ahora el area es de 4. ¿Que nos va a devolver el determinante entonces? Veamoslo

```python
det_A = np.linalg.det(A)
print(det_A)
```

Si tomamos el cuadrado unitario aumento 4 veces su tamaño. De 1 a 4

```python
area_transformada = abs((v1_transformado[0]-v2_transformado[0]))*abs((v1_transformado[1]-v2_transformado[1]))
print(area_transformada)
```

Nos dio exactamente lo mismo. Hagamoslo ahora con un determinante negativo. 

```python
B = A * [-1,1]
print(B)
```

```python
det_B = np.linalg.det(B)
print(det_B) # Va a ser negativo. Porque cuando la matriz es diagonal la determinante
# es la multiplicacion de cada uno de los elementos que están en la diagonal
```

Efectivamente nos devolvió -4

```python
v1 = np.array([0,1])
v2 = np.array([1,0])

v1_transformado = B.dot(v1)
v2_transformado = B.dot(v2)

graficarVectores([v1_transformado,v2_transformado,v1,v2],["blue","red","orange","green"])

plt.xlim(-3,1.5)
plt.ylim(-0.5,2.5)

plt.show()
```

Nuestro espacio rotó. v1 no rotó. Solo se amplificó. v2 se amplifico pero rotó tmb. Está apuntando en la dirección opuesta a la que apunta su v2 original. Entonces nuestra matriz ademas de amplicar su resultado, también invirtio una de sus coordendas. 

Tanto esta modifición como la anterior lo amplificaron en 4, solo que esta ademas lo roto. 

La información del determinante es muy importante pero no nos deja saber la información completa del espacio. Por ejemplo no se si la rotación se da sobre el eje x o sobre el eje y. Si fuese en ambos el detemrinante sería 4 positivo y habria rotado 2 veces.
